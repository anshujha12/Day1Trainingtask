```python
# Histogram

import matplotlib.pyplot as plt
import csv
# load data
data = []
with open(r"C:\Users\Anshu Kumari\Desktop\Mall_Customers.csv", 'r') as file:
    reader = csv.reader(file)
    next(reader) 
    for row in reader:
        data.append(int(row[2]))  
# Create a histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Mall Dataset Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Pie Chart
#It organisize & show data as a % of a whole
# displays data in  a circular-shaped graph
```

```python
import matplotlib.pyplot as plt
import csv
import pandas as pd

# Load data from CSV file using pandas
df = pd.read_csv(r"C:\Users\Anshu Kumari\Desktop\Mall_Customers.csv")

# Count the number of each gender
gender_counts = df['Genre'].value_counts()

# Create a pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%2.1f%%', startangle=50)

# Ensure the pie chart is displayed as a perfect circle
plt.axis('equal')

# Add a title
plt.title('Gender Distribution in the Mall Customer Dataset')

# Show the pie chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r"C:\Users\Anshu Kumari\Desktop\Mall_Customers.csv", encoding="latin1")
df=pd.DataFrame(data)
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Annual Income vs Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```

```python
import matplotlib.pyplot as plt
import csv
import pandas as pd

# Load data from CSV file using pandas
df = pd.read_csv(r"C:\Users\Anshu Kumari\Desktop\Mall_Customers.csv")

# Count the number of each gender
gender_counts = df['Genre'].value_counts()
# Define colors for male and female bars
colors = ['blue', 'orange']

# Create a bar chart with different colors for male and female
plt.bar(gender_counts.index, gender_counts, color=colors)
# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution in the Mall Customer Dataset')

# Show the bar chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file using pandas
df = pd.read_csv(r"C:\Users\Anshu Kumari\Desktop\Mall_Customers.csv")

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['Genre'] == 'Male']['Age'], df[df['Genre'] == 'Female']['Age']], labels=['Male', 'Female'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot of Age Distribution by Gender')

# Show the box plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file using pandas
df = pd.read_csv(r"C:\Users\Anshu Kumari\Desktop\Mall_Customers.csv")

# Create a regression plot using Seaborn
plt.figure(figsize=(8, 6))
sns.regplot(x='Age', y='Annual Income (k$)', data=df)
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Regression Plot of Age vs. Annual Income')

# Show the regression plot
plt.show()
```
